## BEST KAGGLE: NUMERICAL + LOG ##

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import metrics
```

```python
# read in csv
df = pd.read_csv('../datasets/train.csv')
df_test = pd.read_csv('../datasets/test.csv')
pd.options.display.max_columns = 999
df = df.reindex(sorted(df.columns), axis=1)
df_test = df_test.reindex(sorted(df_test.columns), axis=1)
df.drop([960, 1885], inplace=True)
```

## CLEANING / EDA (HAD DONE MUCH IN PREVIOUS ITERATIONS) ##

```python
#check for null
df.isnull().sum().sort_values()
```

```python
#replaced 14 of 26 features with NaN to NA as in data dictionary

df = df.fillna({'Pool QC': 'NA',
           'Misc Feature': 'NA',
           'Alley': 'NA',
           'Fence': 'NA',
           'Fireplace Qu': 'NA',
           'Garage Finish': 'NA',
           'Garage Qual': 'NA',
           'Garage Cond': 'NA',
           'Garage Type': 'NA',
           'Bsmt Exposure': 'NA',
           'BsmtFin Type 2': 'NA',
           'BsmtFin Type 1': 'NA',
           'Bsmt Cond': 'NA',
           'Bsmt Qual': 'NA',
           'Mas Vnr Type': 'NA',
           'Electrical': 'NA',
})
df_test = df_test.fillna({'Pool QC': 'NA',
           'Misc Feature': 'NA',
           'Alley': 'NA',
           'Fence': 'NA',
           'Fireplace Qu': 'NA',
           'Garage Finish': 'NA',
           'Garage Qual': 'NA',
           'Garage Cond': 'NA',
           'Garage Type': 'NA',
           'Bsmt Exposure': 'NA',
           'BsmtFin Type 2': 'NA',
           'BsmtFin Type 1': 'NA',
           'Bsmt Cond': 'NA',
           'Bsmt Qual': 'NA',
           'Mas Vnr Type': 'NA',
           'Electrical': 'NA',
})
```

```python
#check for null again
df.isnull().sum().sort_values()
```

```python
#relpace missing values with mean value
feat_list = ['Lot Frontage', 'Garage Yr Blt', 'Mas Vnr Area', 'Bsmt Full Bath', 'Bsmt Half Bath', 'BsmtFin SF 1', 'Total Bsmt SF', 'Garage Area', 'Garage Cars', 'Bsmt Unf SF', 'BsmtFin SF 2']
for feat in feat_list:
        df[feat].fillna(df[feat].mean(), inplace=True)

test_feat_list = ['Lot Frontage', 'Garage Yr Blt', 'Mas Vnr Area', 'Bsmt Full Bath', 'Bsmt Half Bath', 'BsmtFin SF 1', 'Total Bsmt SF', 'Garage Area', 'Garage Cars', 'Bsmt Unf SF', 'BsmtFin SF 2']
for feat in test_feat_list:
        df_test[feat].fillna(df_test[feat].mean(), inplace=True)
```

```python
#verify that null are 0
df_test.isnull().sum().sum()
```

```python
#select only columns that are num, i.e. not 'object', set equal to variable
num_columns = df.dtypes[df.dtypes != 'object'].index
# select all null, set equal to variable
is_nulls = df.isnull().sum()
# selsct all not null, set equal to variable
not_null_columns = is_nulls[is_nulls==0].index
selected = df[not_null_columns & num_columns]
selected
```

```python
#set X, y
#selected is dataframe, remove 'SalePrice'
X = selected.drop(columns='SalePrice')
#?X = X.transpose()
features = X.columns
y = df['SalePrice']
#instantiate
lr = LinearRegression()
#split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 33)
#log
y_train_log = y_train.map(np.log)
y_test_log = y_test.map(np.log)

cross_val_score(lr, X_train, y_train_log).mean()
```

```python
lr.fit(X_train, y_train_log)

lr.score(X_train, y_train_log)
```

```python
lr.score(X_test, y_test_log)
```

```python
pred = lr.predict(X_test)

residuals = y_test_log - pred
```

 ASSESS:
 - train score = 89.9
 - test score = 89.2
 - cross_val score = 89.3
 looks like low variance, low on bias, LOOKS LIKE A GOOD SCORE!
 89.3% of the variability of the target can be explained in this model 

```python
#scatterplotting residuals 
plt.axhline(y=0, color='r')
sns.scatterplot(x=pred, y=residuals)
('Residual Plot to Show error - Epsilon')
plt.xlabel('predictions')
plt.ylabel('residuals');
```

 - Linear
 
 - (E) Homoscedastic
 
 - Normally distributed
